<h3 style="color:red">Line chart of csv file</h3>

```python
import pandas as pd
iris=pd.read_csv('iris.csv')
print(iris.head())
```

```python
# droping last column 'Species' because we cannot use it in graph and dropping 'Unnamed: 0' aslo we don't need it.

columns = iris.columns.drop(['Species','Unnamed: 0'])

# create x data
x_data = range(0, iris.shape[0])

# create figure and axis

fig, ax = plt.subplots()

# plot each column

for column in columns:
    ax.plot(x_data, iris[column], label=column)
    
# set title and legend

ax.set_title('Iris Dataset')
ax.legend()
# legend = info box
```

<h3 style="color:red">Histogram</h3>
<p>A histogram is a graph that shows the frequency of numerical data using rectangles. The height of a rectangle (the vertical axis) represents the distribution frequency of a variable (the amount, or how often that variable appears).</p>

```python
# create fig and axis

fig, ax = plt.subplots()

# plot histogram

ax.hist(iris['Sepal.Length'])

# set title and legend

ax.set_title('Iris Dataset')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Frequency')
```

<h3 style="color:red">Bar Plot</h3>
<p>Bar charts display categorical data, or information that's separated into different groups based on characteristics, while histograms display numerical or quantitative data in bins, which is data that you can measure with numbers.</p>

```python
wine_reviews = pd.read_csv('winemag.csv')
wine_reviews.head()
```

```python
# Bar chart

fig, ax = plt.subplots()

# count the occurance of each class
data = wine_reviews['points'].value_counts()

# get x and y data

points = data.index
frequency = data.values

# create bar chart

ax.bar(points, frequency)

# set title and labels

ax.set_title('Wine Reviews')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

```python
# BAR Chart by one line

wine_reviews['points'].value_counts().sort_index().plot.bar()
```

```python
# Countries on X-axis and frequencies on the Y-axis

wine_reviews.groupby('country').price.mean().sort_values(ascending=False)[:5].plot.bar()

# [:5] for first five countries data , ascending false because I want graph in drecreasing order.
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (12,7))
# sixe we use to adjust bar size

countries = ['USA','Brazil', 'Russia','Spain', 'UK','Canada']


totalCases = [112596, 137312 , 345638 ,121748, 140283, 148379]
totalDeaths = [11125, 11373 , 13438 ,11748, 11283, 11479]

for i in range(len(countries)):
    plt.bar(countries[i], totalDeaths[i], bottom = totalCases[i] - totalDeaths[i], color = 'red', alpha = 0.8)
    plt.bar(countries[i], totalCases[i] - totalDeaths[i], color = 'yellow', alpha = 0.5)

plt.legend(labels=["Total Deaths","Total Cases"])
plt.title('Total Cases VS Deaths')

plt.ylabel('Deaths')
plt.xlabel('Countries')

    

plt.show()
```

<h3 style="color:red">Pie Chart</h3>

```python
from pandas import DataFrame
import matplotlib.pyplot as plt

Data = {'Tasks' : [300,500,700],
       'Task Type' : ['Tasks Pending','Tasks Ongoing','Task Completed']}

df = DataFrame(Data)
df.set_index('Task Type', inplace=True)

df.plot.pie(y='Tasks', figsize=(10,10),autopct='%1.1f%%', startangle=90)

# autopct enables you to display the percent value using Python string formatting.
# autopct='%1.1f%%' means 1.1f float num.

```

```python
import numpy as np
import matplotlib.pyplot as plt

# if using jupyter notebook , include:
%matplotlib inline

labels = ['Civil','Electrical','Mechanical','Chemical']
sizes = [15,20,45,10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')

ax.axis('equal')
ax.set_title('Engineering Diciplines')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# if using jupyter notebook , include:
%matplotlib inline

labels = ['Civil','Electrical','Mechanical','Chemical']
sizes = [15,20,45,10]

explode = (0.1,0.1,0.1,0.4)
# we use explode to seperate each part from origine

fig, ax = plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='%1.1f%%',
       shadow = True,
       startangle = 90)

ax.axis('equal')
ax.set_title('Engineering Diciplines')

plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.bar(range(1,6), np.random.randint(1,20,5), color = 'r', width = 0.5, alpha = 0.5)
plt.title("2,2,1 = first graph")

plt.subplot(2,2,2)
plt.bar(range(1,6), np.random.randint(1,20,5), color = 'orange', width = 0.25, alpha = 0.5)
plt.title("2,2,2 = Second graph")

plt.subplot(2,2,3)
plt.scatter(range(1,6), np.random.randint(1,20,5) , s=100 , color = "g")
plt.title("2,2,3 = third graph")

plt.subplot(2,2,4)
plt.plot(range(1,6), np.random.randint(1,20,5) , marker = '*' , color = "red", linestyle="--")
# marker = 'o' point style , see last plot
plt.title("2,2,4 = fourth graph")


```
