<h1 style="color:red">Uniform Distribution</h1>

```python
from scipy.stats import randint
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1)

# Calculate a few first moments:
# need max and min range just like a , b 

low , high = 7, 31
mean, var, skew, kurt = randint.stats(low, high, moments='mvsk')
# moments='mvsk' for mean, var, skew, kurt

# Displaying Probability Mass Function(PMF)

x= np.arange(randint.ppf(0.01, low, high),randint.ppf(0.99, low, high))
# ppf = percent point function(convert value 0.01 to 0.99)

ax.plot(x, randint.pmf(x, low, high), 'bo', ms=8 , label='randint pmf')
ax.vlines(x, 0, randint.pmf(x, low, high),colors='r', lw=5, alpha=0.5)

# Alternatively , the distribution object can be called (as a function)
# to fix the shape and location. This return a "frozen" RV object holding
# the given parameters fixed

# freeze the distribution and display the frozen 'pmf':

rv = randint(low ,high)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1, label='frozen pmf')
ax.legend(loc='best', frameon=False)
# frameon=False because both graph on same
plt.show()

# Check occuracy of 'CDF' and 'PPF':

prob = randint.cdf(x , low, high)
np.allclose(x, randint.ppf(prob, low, high))
# True

# Generate random numbers:

r = randint.rvs(low, high, size=1000)
```

<h1 style="color:purple">Bernoulli Distribution</h1>

**Bernoulli Distribution Definition**

A discrete probability distribution wherein the random variable can only have 2 possible outcomes is known as a Bernoulli Distribution. If in a Bernoulli trial the random variable takes on the value of 1, it means that this is a success. The probability of success is given by p. Similarly, if the value of the random variable is 0, it indicates failure.

```python
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size = 10000, p =0.6)

# rvs = Random Variates
```

```python
ax = sns.distplot(data_bern,
                 kde = False,
                 color = 'skyblue',
                 hist_kws = {"linewidth" : 15, 'alpha' : 1})

ax.set(xlabel = "Bernoulli Distribution", ylabel = "Frequency")
```

<h1 style="color:purple">Binomial Distribution</h1>

**Binomial Distribution Definition**

The binomial distribution is a probability distribution used in statistics that states the likelihood that a value will take one of two independent values under a given set of parameters or assumptions.


```python
from scipy.stats import binom
data_binom = binom.rvs(n = 10, p = 0.8, size = 10000)
```

```python
ax = sns.distplot(data_binom,
                 kde = False,
                 color = 'skyblue',
                 hist_kws = {"linewidth": 15, "alpha" : 1})

ax.set(xlabel = 'Binomial Distribution' , ylabel = 'Frequency')
```

<h1 style="color:purple">Poisson Distribution</h1>

**Poisson Distribution Definition**

The Poisson distribution is a discrete probability function that means the variable can only take specific values in a given list of numbers, probably infinite. A Poisson distribution measures how many times an event is likely to occur within x period of time. In other words, we can define it as the probability distribution that results from the Poisson experiment. A Poisson experiment is a statistical experiment that classifies the experiment into two categories, such as success or failure. Poisson distribution is a limiting process of the binomial distribution.

```python
from scipy.stats import poisson
data_poisson = poisson.rvs(mu = 3, size = 10000)
```

```python
ax = sns.distplot(data_poisson,
                 kde = False,
                 color = 'skyblue',
                 hist_kws = {"linewidth": 15, "alpha" : 1})

ax.set(xlabel = 'Poisson Distribution' , ylabel = 'Frequency')
```

<br><b>
A warehouse typically receives 8 deliveries between 4 and 5 pm on Friday <br>

1. what is the Probability that only 4 deliveries will arrives between 4 and 5 pm on Friday?<br>
2. What is the Probability of having less than 3 deliveries on Friday between 4 and 5 pm ? <br>
3. What is the Probability of having no deliveries on Friday between 4 and 5 pm ?
</b>

```python
# 1. what is the Probability that only 4 deliveries will arrives between 4 and 5 pm on Friday?

from scipy.stats import poisson
poisson.pmf(4, 8)

# pmf(4, 8) = because only 4 deliveries out of 8
```

```python
# 2. What is the Probability of having less than 3 deliveries on Friday between 4 and 5 pm ?

from scipy.stats import poisson
poisson.cdf(3, 8)

# cdf(3, 8) = because less than 3 means from 0 to 3
```

```python
# 3. What is the Probability of having no deliveries on Friday between 4 and 5 pm ?

from scipy.stats import poisson
poisson.pmf(0, 8)
```
